```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
rd = pd.read_csv('online_retail.csv', encoding='utf-8-sig')
```

```python
print(rd.columns)
```

```python
rd.head()
```

```python
# Clean column names
rd.columns = rd.columns.str.strip()
```

```python
print(rd.columns)
```

```python
print("DataFrame columns:", rd.columns.tolist())
```

```python
# Convert datetime in the correct format
rd['InvoiceDate'] = pd.to_datetime(rd['InvoiceDate'], dayfirst=True)
#rd['InvoiceDate'] = rd['InvoiceDate'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
rd.head()
```

```python
# Get the date range
min_date = rd['InvoiceDate'].min()
max_date = rd['InvoiceDate'].max()
date_range = max_date - min_date
    
print(f"Start date: {min_date}")
print(f"End date: {max_date}")
print(f"Date range: {date_range}")
print(f"Total days: {date_range.days}")
```

```python
# Getting the basic statistics of the numerical columns of the dataset
rd.describe()
```

```python
#Checking for missing values
rd.isnull().sum()
```

```python
# We found in Excel that the Description column includes many characters like ?, we need to remove these characters
#rd['Description'] = rd['Description'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
# Now we need to check for missing values again
rd.isnull().sum()
```

```python
# We will now remove the misssing values 
rd = rd.dropna(subset=['CustomerID', 'Description'])
```

```python
rd.isnull().sum()
```

```python
# Saving the cleaned data
rd.to_csv("clean_retail.csv", index=False)
```
