## Making the Category feature
After initially visualizing the data in PowerBI, I discovered that the charts weren't very intuitive due to a lack of higher-level categorisation. I had all the names of the individual products sold by the business but not the categories that they belonged to, which resulted in messy charts and information overload. Hence, I decided to create a category column using the method below to solve that issue and allow for better data visualization.

```python
import pandas as pd

df = pd.read_csv("clean_retail.csv", encoding='utf-8-sig')
```

```python
df['Revenue'] = df['Quantity']*df['UnitPrice']
```

```python
df['Description'] = df['Description'].str.lower().str.strip()
```

```python
#  Improved
def categorize_product(desc):
    if pd.isna(desc):
        return "Other"
    
    desc = desc.lower()
    
    if any(word in desc for word in ["cake", "bake", "kitchen", "plate", "cup", "mug"]):
        return "Kitchen"
    
    elif any(word in desc for word in ["bag", "purse", "wallet"]):
        return "Accessories"
    
    elif any(word in desc for word in ["lamp", "light", "lantern"]):
        return "Lighting"
    
    elif any(word in desc for word in ["basket", "box", "storage", "tin"]):
        return "Storage"
    
    elif any(word in desc for word in ["christmas", "gift", "wrap", "card"]):
        return "Seasonal & Gifts"
    
    elif any(word in desc for word in ["heart", "decor", "frame", "mirror", "vintage"]):
        return "Home Decor"
    
    elif any(word in desc for word in ["toy", "kids", "doll", "game"]):
        return "Kids & Toys"
    
    elif any(word in desc for word in ["garden", "outdoor"]):
        return "Outdoor"
    
    else:
        return "Other"
```

```python
df['Category'] = df['Description'].apply(categorize_product)
```
